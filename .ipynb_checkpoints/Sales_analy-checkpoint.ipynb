```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fl="Sample - Superstore.xls"
df = pd.read_excel(fl, engine='xlrd')

df.head()
```

```python
sales=df['Sales'].to_numpy()
qty=df['Quantity'].to_numpy()
category=df['Category'].to_numpy()
region=df['Region'].to_numpy()

print(f"Total Records:{len(df)}")
print(f"Columns:{df.columns.tolist()}")
```

```python
sum_sales = np.sum(sales)
print(f"Total Sales: ${sum_sales:.2f}")

avg = np.mean(sales)
print(f"Average Sales per Transaction: ${avg:.2f}")

var = np.var(sales)
print(f"Sales Variance: ${var:.2f}")

total_qty = np.sum(quantity)  # Corrected variable name from `qty` to `quantity`
print(f"Total Quantity Sold: {total_qty}")

unique_category, count_category = np.unique(category, return_counts=True)
print("Sales Count by Category:")
for cat, count in zip(unique_category, count_category):
    print(f"{cat}: {count}")

unique_region = np.unique(region)
reg_avg_sales = {reg: np.mean(sales[region == reg]) for reg in unique_region}  # Corrected to use boolean indexing
print("Average Sales per Region:")
for reg, avg_sales in reg_avg_sales.items():
    print(f"{reg}: ${avg_sales:.2f}")

```

```python
plt.figure(figsize=(8, 5))
plt.bar(unique_category, count_category, color='skyblue')
plt.title('Sales Count by Category')
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(unique_region, [reg_avg_sales[reg] for reg in unique_region], color='salmon')
plt.title('Average Sales per Region')
plt.xlabel('Region')
plt.ylabel('Average Sales ($)')
plt.show()
```
